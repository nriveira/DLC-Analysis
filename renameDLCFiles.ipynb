```python
## Code to check if the renaming process works properly.
import os

directory = r"C:\Users\nrive\Research\AnkG\DLC_Vids\iteration1\Mouse-19"

for (dirpath, _ , filenames) in os.walk(directory):
    for filename in filenames:
        if filename.endswith('.avi'): 
            filePath = dirpath + '\\' + filename
            
            splitDir = dirpath.split('\\')
            parentDir = ''
            for n in splitDir[:7]:
                parentDir += n + '\\'
            
            dirName = ''.join(splitDir[7:]) 
            destPath = parentDir + dirName + filename
            
            print(filePath + ':\n' + destPath + '\n')
```

```python
## Code to move all files to the same directory.
for (dirpath, _ , filenames) in os.walk(directory):
    for filename in filenames:
        if filename.endswith('.avi'): 
            filePath = dirpath + '\\' + filename
            
            splitDir = dirpath.split('\\')
            parentDir = ''
            for n in splitDir[:7]:
                parentDir += n + '\\'
            
            dirName = ''.join(splitDir[7:]) 
            destPath = parentDir + dirName + filename
            
            os.rename(filePath, destPath)
```
